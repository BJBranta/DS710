For Boolean expressions, `&` and `and` represent logical "and": `True & True` returns `True`, but `True & False`
returns `False`, because at least one of the arguments is not true. Meanwhile, `|` and `or` represent logical "or": `True | True` and `True | False` both return `True`, because at least one of the arguments is true in each case. 

🧠 Predict the outcome of the following segments of code, then run the code to check your reasoning.

```python
x = 12
if (x > 9) & (x % 2 != 0):
    print('8')
else:
    print('11')
```

```python
x = 12
if (x > 9) | (x % 2 != 0):
    print('8')
else:
    print('11')
```

Notice that `if x:` interprets $x$ as a Boolean with value `True` if the value assigned to $x$ is not `0` nor `None`. However, the Boolean `y==True` is `False` even when the value assigned to $y$ is $6$; one should avoid such a statement.

```python
x=None
y=6
if x:
    print(2)
if y:
    print(5)
if y==True:
    print(8)
```

Also be aware that `and`/`&` and `or`/`|` can work in unexpected ways when not applied to Booleans, since one is a Boolean operator and one is a bitwise operator. For example:

```python
x=2
y=5
if x and y:
    print('x and y is True')
if x & y:
    print('x & y is True')
```

Predict the difference in outputs of the following two `for` loops, then run the code to check your reasoning.

Python allows us to iterate over elements of different data types. Below, we iterate over characters in a string.

```python
for c in "abcdefghijklmnop":
    print(f'{c} is for {4*c}!')
```

We can exit a loop using the `break` command. 🧠 Predict the output of and then run the code below.

```python
for i in range(10):
    if i*i == 25: 
        break
    print(i*i)
```

We can skip a particular iteration using the `continue` command. 🧠 Predict the output of and then run the code below.

### FizzBuzz

🧠 Using what you have learned, write a program that prints the integers from $1$ to $100$ with the following exceptions: (1) for multiples of five (but not seven), print "Fizz" instead of the number, (2) for the multiples of seven (but not five), print "Buzz", and (3) for numbers which are multiples of both five and seven, print "FizzBuzz".

```python
for ii in range(1, 1010, 1):
    if (ii % 5 == 0) and (ii % 7 != 0):
        print('Fizz')
        print(ii)
    elif (ii % 5 != 0) and (ii % 7 == 0):
        print('Buzz')
        print(ii)
    elif (ii % 5 == 0) and (ii % 7 == 0):
        print('FizzBuzz')
        print(ii)
```
