We can concatenate two lists or tuples using `+`; to add to an existing list, we can alternatively (and preferably, for long lists) use `extend`. We access elements of a list or tuple using square brackets `[]`.

```python
plus_list = [3, 8, 10, 2] + [2, 7, 5]
print(plus_list)
plus_list.extend([4.5, 3, 7])
print(plus_list)
print(plus_list[0])

plus_tuple = (5, 7, 5) + (['apples','bananas'], 3.1415)
print(plus_tuple)
print(plus_tuple[3])
```

We can add single elements to the end of a list (but not a tuple!) using `append`.

```python
threes = []
threes.append(3)
print(threes)
threes.append(3)
print(threes)
threes.append(3)
print(threes)
```

```python
threes_tuple = ()
threes_tuple.append(3)
```

We can count the instances of a value in a tuple or list using `count`.

```python
print(plus_tuple.count(5))
print(threes.count(3))
```

### Dictionaries 

##### (see <a href="https://www.python-course.eu/python3_dictionaries.php">here</a> for an additional resource on dictionaries)


The two lists below contain years and the corresponding best-selling albums in the United States according to *Billboard* (<a href="https://en.wikipedia.org/wiki/List_of_best-selling_albums_by_year_in_the_United_States">source</a>).

```python
years=[1956, 1966, 1976, 1986, 1996, 2006, 2016]
albums=['Calypso', 'Whipped Cream & Other Delights', 'Frampton Comes Alive!', 
       'Whitney Houston', 'Jagged Little Pill', 'High School Musical', 'Views']
print(years)
print(albums)
```

These two lists clearly pair nicely. Moreover, it makes more sense to access album titles by year than by offset. Let's make a dict from the two lists.

```python
bestsellers = dict(zip(years, albums))
print(bestsellers)
```

🧠 Predict what the code below will output, then run the code.

```python
bestsellers[2]
```

Recall that elements of a dict are accessed using their keys, not offsets. We can check if a dict contains a key:

```python
1970 in bestsellers
```

```python
1966 in bestsellers
```

```python
print(bestsellers.keys())
print(bestsellers.values())
```

We may wish to add entries to our dict:

```python
bestsellers[1970] = 'Bridge over Troubled Water'
print(bestsellers)

bestsellers.update({1960: 'The Sound of Music', 1980: 'The Wall', 1990: 'Janet Jackson\'s Rhythm Nation 1814', 
                    2000: 'No Strings Attached', 2010: 'Recovery'})
print(bestsellers)
```

Or delete them:

```python
bestsellers.pop(1966)
```

```python
del bestsellers[2010]
print(bestsellers)
print(bestsellers.get(2010))
```

Suppose we wish to print `[album] was the bestseller of [year].` for each of the ablum-year pairs. We could loop over both lists:

```python
for alb in albums:
    for yr in years:
        if years.index(yr) == albums.index(alb):
            print(f'{alb} was the bestseller of {yr}.')
```

Above, we looped first over a list of albums, then over a list of years. To be more efficient, we can use the dictionary that links the two, rewriting the code above so that it loops once over the dict.

```python
bestsellers=dict(zip(years, albums))

for yr in bestsellers:
    print(f'{bestsellers[yr]} was the bestseller of {yr}.')
```

Recall the `for` loop we implemented above to create a list of the integers between 0 and 50, inclusive, that are integer multiples of 10.

```python
ten_divides=[]
for i in range(51):
    if i%10 == 0:
        ten_divides.append(i)
        
print(ten_divides)
```

Using list comprehensions, we can simplify this loop into a single line of code.

```python
ten_divides_comp = [i for i in range(51) if i%10 == 0]

print(ten_divides_comp)
```
