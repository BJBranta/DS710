### A set of diamonds


The data set `diamonds.txt` contains data on the characteristics of diamonds and their prices.  There are 5 levels of clarity represented in the data:  from clearest to least clear, these levels are IF, VVS1, VVS2, VS1, and VS2, where

* IF = Internally Flawless
* VVS = Very, very slight inclusions
* VS = Very slight inclusions

We would like to write a function to count the number of diamonds of each type.  Later in the course, we will discuss how to use libraries for this type of analysis.  For now, we will ignore much of the structure of `diamonds.txt`, and just treat it as a text file without any additional organization. First, we will read the data in `diamonds.txt` as a string. Of course, you'll need to have ```diamonds.txt``` in the same directory as this notebook to run the demo.

````python
with open("diamonds.txt") as diamond_file:
    diamond_string = diamond_file.read()
````

Next, we define the following first attempt at a counting function:

````python
def clarity_count_v1(s, clarity):
    """Return the number of instances of the indicated 
    diamond clarity in the given string.
    
    Parameters
    ----------
    s: str
        A string containing the description of some diamonds.
    
    clarity: str
        A string of a clarity type: 'IF', 'VVS1', 'VVS2', 'VS1', or 'VS2'
    
    Returns
    -------
    clar_count: int
        Number of instances of diamond clarity.
    """
    clar_count = s.count(clarity)
    return clar_count
````

### Improving the function

Let's improve `clarity_count` in two ways: we will add a default clarity of 0, and we will display an error message if the specified clarity number is not an integer between 0 and 4. Note that Python incorporates sophisticated error-checking functionality.  The method we use here is appropriate for informal data analysis, but not for production code.

````python
def improved_clarity_count(s, clarity='IF'):
    """Return the number of instances of the indicated 
    diamond clarity in the given string.
    
    Parameters
    ----------
    s: str
        A string containing the description of some diamonds.
    
    clarity: str
        A string of a clarity type: 'IF', 'VVS1', 'VVS2', 'VS1', or 'VS2'
    
    Returns
    -------
    clar_count: int
        Number of instances of diamond clarity.
    """
    if type(clarity)!= str:
        raise TypeError('Warning: clarity must be an string.')
        return
    
    valid_clarities = ['IF', 'VVS1', 'VVS2', 'VS1','VS2']
    if not clarity in valid_clarities:
        raise ValueError("Warning: clarity must be one of 'IF', 'VVS1', 'VVS2', 'VS1', or 'VS2'")
        return
    
    clar_count = s.split().count(clarity)
    return clar_count
````

### Writing Data 

In the example above, we read data from the diamonds file.  Let's also write the data of the counts.


*Lambda functions* are anonymous functions where we do engage in the entire function definition machinery.  Lambda functions can be useful if we want to apply a short function to all elements of a list or other data structure. To demonstrate lambdas, we will use it to define a way to format lines that we will write to a text file.

Note that normally, lambda functions are not used in this way, but instead when we want to apply a function repeatedly without naming it.  For the purpose of illustrartion we name it in this example.

````python
line_formatter = lambda clarity: f"{clarity}: {improved_clarity_count(diamond_string,clarity)}\n"

for clarity in ['IF', 'VVS1', 'VVS2', 'VS1','VS2']:
    with open("clarity_counts.txt",'w') as output_file:
        output_file.write(line_formatter(clarity))
````

Note that running this line will result in a file with only a single line.  This is because using the `'w'` argument in open means that each time we `open` the file it overwrites what's there.  This isn't by default a bad thing, we don't want to add more and more copies of the same information when we run the code repeatedly.

### Currying


Now, suppose we care about the percentage of diamonds in our dataset that have the lowest clarity: VS2. We can use currying and our function `improved_clarity_count` to determine this percentage. In particular, the first argument from `improved_clarity_count` is curried to create `percent_not_clear`.

````python
clarities = ['IF', 'VVS1', 'VVS2', 'VS1','VS2']
def percent_not_clear(s):
    numerator = 100*improved_clarity_count(s,"VS2")
    denominator = sum([improved_clarity_count(s,c) for c in clarities])
    return numerator/denominator
````

````python
percent_not_clear('VVS1 VVS1 VS1 VVS1 VS1 IF IF VS2')
````

````python
percent_not_clear(diamond_string)
````

### Nested functions


Now, suppose we wish to define a function that gives percentages rather than counts. Assuming we are starting from scratch, we can use a <em>nested function</em> so that a count function is defined within the definition of the percentage function. Observe:

````python
def clarity_percentage(s):
    """Return the percentage of each of the five diamond clarities in a list format. 
    
    Parameters
    ----------
    s: str
        A string containing the description of some diamonds.
    
    Returns
    -------
    clar_per: list
        List of percentages of diamonds of each clarity.
    """
    def _clarity_count(s, clarity='IF'):
        """Return the number of instances of the indicated 
        diamond clarity in the given string.
        
        Parameters
        ----------
        s: str
            A string containing the description of some diamonds.
        
        clarity: str
            A string of a clarity type: 'IF', 'VVS1', 'VVS2', 'VS1', or 'VS2'
        
        Returns
        -------
        clar_count: int
            Number of instances of diamond clarity.
        """
        if type(clarity)!= str:
            raise TypeError('Warning: clarity must be an string.')
            return
        
        valid_clarities = ['IF', 'VVS1', 'VVS2', 'VS1','VS2']
        if not clarity in valid_clarities:
            raise ValueError("Warning: clarity must be one of 'IF', 'VVS1', 'VVS2', 'VS1', or 'VS2'")
            return
        
        clar_count = s.split().count(clarity)
        return clar_count
        
    valid_clarities = ['IF', 'VVS1', 'VVS2', 'VS1','VS2']
    clar_per = [round(100*_clarity_count(s,c)/sum([_clarity_count(s,d) for d in valid_clarities]),2) for c in valid_clarities]
    return clar_per
````

````python
clarity_percentage('VVS1 VVS1 VS1 VVS1 VS1 IF IF VS2')
````

````python
clarity_percentage(diamond_string)
````
