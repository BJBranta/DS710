### Beginner fizzbuzz

#### Background of fizzbuzz 

According to internet lore, a common coding interview question from the 2010's was [Fizzbuzz](https://en.wikipedia.org/wiki/Fizz_buzz).  The interviewer would ask the candidate to write code accomplishing a simple task.  It's essentially an easy problem that they would expect any programmer to be able to solve immediately without reference or assistance.  

Here's the computational version of the game we'll play in this assignment.  Suppose you have an integer variable `n`.  Then let `n`'s "fizzbuzz number" be `n_fb` be the string according to the following chart:

* `n_fb` is `'fizz'` if `n` is divisible by `3`
* `n_fb` is `'buzz'` if `n` is divisible by `5`
* `n_fb` is `'fizzbuzz'` if `n` is divisible by `3` and `5`
* `n_fb` is `''` (the empty string) if `n` is not divisible by 3 or 5

You can read more about fizzbuzz on [the linked wikipedia page](https://en.wikipedia.org/wiki/Fizz_buzz).  

Below you will see a function that partially complete this task with some mistake in the control-flow. We'll get back to that in a minute, but first take a look at the function.

Note that The function, `fizzbuzz`, as written uses several different `return` calls. The moment a function `return`s, the rest of it will not be run. However, if the condition to hit that `return` is not triggered, it is not used.  

```python
def fizzbuzz(n):
    if n%3 == 0:
        return 'fizz'
    elif n%5 == 0:
        return 'buzz'
    elif n%3 == 0 and n%5 == 0:
        return 'fizzbuzz'
    else:
        return ''
```

Because we have a function here, we can quickly test the accuracy of the code for many different values of `n`.

### Washington, D.C. metro fare calculation


The Washington Metropolitan Area Transit Authority (WMATA) calculates metro fares according to the distance traveled and time of day (source: <a href="https://www.wmata.com/fares/smartrip/index.cfm">WMATA SmarTrip</a> and <a href="https://www.wmata.com/about/records/public_docs/upload/Tariff-39-on-Rates-Fares-Annotated-Final-6-28-19.pdf">fare details</a>). Suppose WMATA wants to develop a Python function that will calculate a customer's fare. As a first attempt, your colleague has defined the following function. Read through the function definition and try to understand how WMATA calculates metro fares. Your colleague has used the `numpy` <a href="https://numpydoc.readthedocs.io/en/latest/format.html#docstring-standard">standard for documentation strings</a>.

```python
def fare_calc(total_mi, crow_mi, hour):
    """Calculates the cost of a metro ride.
    
    Parameters
    ----------
    total_mi : float
        The shortest possible distance traveled between 
        entrance and exit stations in miles.
    
    crow_mi : float
        The as-the-crow-flies distance between entrance 
        and exit stations in miles.
    
    hour : {'peak', 'off-peak'}
        The time of day.
    
    Returns
    -------
    cost : float
        Cost of the metro ride.
    """
    #Calculate an un-rounded cost if peak hours
    if hour == 'peak':
        #Start with the base peak cost
        cost = 2.25
        #Add the additional cost per composite mile when total > 3 but <= 6
        if 3 < (total_mi+crow_mi)/2 <= 6:
            cost += ((total_mi+crow_mi)/2-3)*0.326
        #Add the additional cost per composite mile when total > 6
        elif (total_mi+crow_mi)/2 > 6:
            cost += 3*0.326 + ((total_mi+crow_mi)/2-6)*0.288
        #The price cap on a peak ride is $6
        cost = min(6, cost)
    #Calculate an un-rounded cost if off-peak hours
    elif hour == 'off-peak':
        #Start with the base off-peak cost
        cost = 2.00
        #Add the additional cost per composite mile when total > 3 but <= 6
        if 3 < (total_mi+crow_mi)/2 <= 6:
            cost += ((total_mi+crow_mi)/2-3)*0.244
        #Add the additional cost per composite mile when total > 6
        elif (total_mi+crow_mi)/2 > 6:
            cost += 3*0.244 + ((total_mi+crow_mi)/2-6)*0.216
        #The price cap on an off-peak ride is $3.85
        cost = min(3.85, cost)
    #Round the cost to the nearest 0.05
    round5 = (100*cost)%5
    if round5 < 2.5:
        cost = (cost*100 - round5)/100
    elif round5 >= 2.5:
        cost = (cost*100 + (5-round5))/100
    #Output that rounded cost
    print(cost) 
```

Suppose that Hyun-Joo has a dict where the keys are pairs of stations he frequently travels between, and the values are tuples containing the relevant distances between them (in 2020), as below.

```python
hyun_joo_stations = {('Silver Spring', 'Shaw-Howard U'): (5.52,5.4), ('Silver Spring', 'College Park'): (8.42,5.6)}
```

He realizes that it would be useful for customers to be able to calculate the peak and off-peak fares between their frequently-used stations, and so writes the following function that will return a dict that has as keys pairs of stations, and as values peak/off-peak fares between those stations.

```python
def freq_fares(freq_travel):
    """Calculates peak and off-peak fares between multiple pairs of stations.
    
    Parameters
    ----------
    freq_travel : dict
        Keys are tuples of pairs of stations,
        values are tuples of total and as-the-crow flies 
        distances between those stations.
    
    Returns
    -------
    freq_travel : dict
        Keys are tuples of pairs of stations, 
        values are tuples of peak/off-peak fares 
        between those stations. 
    """
    for stations in freq_travel.keys():
        freq_travel[stations] = (f"peak: ${fare_calc(freq_travel[stations][0], freq_travel[stations][1], 'peak'):.2f}",
                               f"off-peak: ${fare_calc(freq_travel[stations][0], freq_travel[stations][1], 'off-peak'):.2f}") 
    return freq_travel
```

🧠 Predict what the following code will output, then run it.

```python
hyun_joo_stations =  {('Silver Spring', 'Shaw-Howard U'): (5.52,5.4), ('Silver Spring', 'College Park'): (8.42,5.6)}
print(hyun_joo_stations)
freq_fares(hyun_joo_stations)
print(hyun_joo_stations)
```

Notice that because `dict` is not a literal type, calling the function `freq_fares` actually modifies its argument, `hyun_joo_stations`. This could be problematic, since it is likely that Hyun-Joo actually wants to retain his information about distances: he may want to construct a new dict instead of modifying `freq_travel` in his definition of `freq_fares`. In other scenarios, modifying the argument may be exactly what we want.

Regardless, this will not occur with literals: consider the following simple example, where we see that the value of the int $x$ is not changed by calling `plus_5`.
